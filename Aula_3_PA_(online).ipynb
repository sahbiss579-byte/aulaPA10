```
# Leitura do salário
salario_atual = float(input("Digite o salário atual do funcionário: R$ "))

# Cálculo do aumento (10%)
aumento = salario_atual * 0.10
novo_salario = salario_atual + aumento

# Exibição dos resultados
print(f"O valor do aumento é: R$ {aumento:.2f}")
print(f"O novo salário é: R$ {novo_salario:.2f}")
```

```
# Entrada de dados
base = float(input("Digite o valor da base: "))
altura = float(input("Digite o valor da altura: "))

# Cálculo da área
area = base * altura

# Resultado
print(f"A área do retângulo é: {area} unidades de medida.")

```

```
# Entrada de dados
distancia = float(input("Digite a distância percorrida (km): "))
combustivel = float(input("Digite a quantidade de combustível gasta (litros): "))

# Cálculo do consumo médio
if combustivel > 0:
    consumo_medio = distancia / combustivel
    print(f"O consumo médio do veículo é de: {consumo_medio:.2f} km/l")
else:
    print("Erro: O combustível gasto deve ser maior que zero.")
```
