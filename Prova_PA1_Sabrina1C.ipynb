```python
# Par ou Ímpar

# Entrada
numero = int(input("Digite um número inteiro: "))

# Processamento e Saída
if numero % 2 == 0:
    print(f"O número {numero} é PAR.")
else:
    print(f"O número {numero} é ÍMPAR.")
```

```python
# Maior de Dois Números

# Entrada
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Processamento e Saída
if num1 > num2:
    print(f"O primeiro número ({num1}) é maior que o segundo número ({num2}).")
elif num2 > num1:
    print(f"O segundo número ({num2}) é maior que o primeiro número ({num1}).")
else:
    print(f"Os dois números ({num1} e {num2}) são iguais.")
```

```python
# Aprovação de Aluno

# Entrada
nota = float(input("Digite a nota do aluno: "))

# Processamento e Saída
if nota >= 7:
    print("Status: Aprovado")
elif nota >= 5 and nota < 7:
    print("Status: Recuperação")
else:
    print("Status: Reprovado")
```

```python
# Classificação de Idade

# Entrada
idade = int(input("Digite a idade: "))

# Processamento e Saída
if 0 <= idade <= 12:
    print("Classificação: Criança")
elif 13 <= idade <= 17:
    print("Classificação: Adolescente")
elif 18 <= idade <= 59:
    print("Classificação: Adulto")
else:
    print("Classificação: Idoso")
```

```python
# Senha Correta

# Senha fixa definida no código
senha_fixa = "sabrina1CetecMCM"

# Entrada
senha_digitada = input("Digite a senha: ")

# Processamento e Saída
if senha_digitada == senha_fixa:
    print("Senha Correta!")
else:
    print("Senha Incorreta.")
```

```python
# Classificação de Número (Positivo, Negativo ou Zero)

# Entrada
numero = float(input("Digite um número: "))

# Processamento e Saída
if numero > 0:
    print(f"O número {numero} é POSITIVO.")
elif numero < 0:
    print(f"O número {numero} é NEGATIVO.")
else:
    print(f"O número {numero} é ZERO.")
```

```python
# Calculadora Simples

# Entrada
a = float(input("Digite o primeiro número: "))
b = float(input("Digite o segundo número: "))
operacao = input("Digite a operação (+, -, *, /): ")

# Processamento e Saída
resultado = None
if operacao == '+':
    resultado = a + b
    print(f"O resultado de {a} + {b} é igual a: {resultado}")
elif operacao == '-':
    resultado = a - b
    print(f"O resultado de {a} - {b} é igual a: {resultado}")
elif operacao == '*':
    resultado = a * b
    print(f"O resultado de {a} * {b} é igual a: {resultado}")
elif operacao == '/':
    if b != 0:
        resultado = a / b
        print(f"O resultado de {a} / {b} é igual a: {resultado}")
    else:
        print("Erro: Divisão por zero não é permitida.")
else:
    print("Operação inválida. Por favor, use +, -, * ou /.")
```

```python
# Dia da Semana

# Entrada
numero_dia = int(input("Digite um número de 1 a 7: "))

# Processamento e Saída
dias_da_semana = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"
}

if numero_dia in dias_da_semana:
    print(f"O dia correspondente ao número {numero_dia} é: {dias_da_semana[numero_dia]}.")
else:
    print("Número inválido. Por favor, digite um número entre 1 e 7.")
```

```python
# Ano Bissexto

# Entrada
ano = int(input("Digite um ano: "))

# Processamento e Saída
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é um ano BISSEXTO.")
else:
    print(f"O ano {ano} NÃO é um ano bissexto.")
```

```python
# Desconto em Compras

# Entrada
valor_compra = float(input("Digite o valor da compra: "))

# Processamento e Saída
disconto = 0
if valor_compra >= 100:
    disconto = 0.10  # 10% de desconto
elif valor_compra >= 50 and valor_compra < 100:
    disconto = 0.05  # 5% de desconto

valor_final = valor_compra * (1 - disconto)

print(f"Valor da compra: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: {disconto * 100:.0f}%")
print(f"Valor final a pagar: R$ {valor_final:.2f}")
```
