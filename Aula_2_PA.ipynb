```
# Calcular a conversão de Celsius para Fahrenheit

# Entrada
celsius = float(input("Digite a temperatura em graus Celsius: "))

# Processamento
fahrenheit = (celsius * 9/5) + 35

# Saída
print(f"A temperatura em Fahrenheit é igual: {fahrenheit}")

```
