```
# Escrever um programa que faça calculos aritmeticos

# Entrada

a = int(input("digite o primeiro número: "))
b = int(input(" digite o primeiro número: "))

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b # Usar "//" para divisão inteira
resto = a % b # o simbolo de % é igual ao resto da divisão

# Saída

print ("|--------------------CALCULADORA SIMPLES--------------------|")
print ("|                                                           |")
print ("|---------------------------soma----------------------------|")
print ("|                                                           |")
print (f"o resultado da soma de {a} mais {b} é igual: {soma}")
print ("|                                                           |")
print ("|-------------------------subtração-------------------------|")
print ("|                                                           |")
print (f"o resultado da subtração de {a} menos {b} é igual: {subtracao}")
print ("|                                                           |")
print ("|----------------------multiplicação------------------------|")
print ("|                                                           |")
print (f"o resultado da multiplicação de {a} vezes {b} é igual: {multiplicacao}")
print ("|                                                           |")
print ("|----------------------divisão inteira----------------------|")
print (f"o resultado da divisão de {a} dividido {b} o resto é igual: {div_int}")
print ("|                                                           |")
print ("|---------- -----------resto da divisão---------------------")
print ("|                                                           |")
print (f"o resultado da divisão de {a} dividido {b} o resto é igual: {resto}")
```

```
# Entrada
num = int(input("Digite um número inteiro: "))

# Processamento
dobro = num * 2
metade = num / 2
resto = num % 3

# Saída
print(f"O dobro de {num} é: {dobro}")
print(f"A metade de {num} é: {metade}")
print(f"O resto da divisão de {num} por 3 é: {resto}")
```

```
# Calcular a conversão de Celsius para Fahrenheit

# Entrada
celsius = float(input("Digite a temperatura em graus Celsius: "))

# Processamento
fahrenheit = (celsius * 9/5) + 35

# Saída
print(f"A temperatura em Fahrenheit é igual: {fahrenheit}")
```

```
# Leitura do salário
salario_atual = float(input("Digite o salário atual do funcionário: R$ "))

# Cálculo do aumento (10%)
aumento = salario_atual * 0.10
novo_salario = salario_atual + aumento

# Exibição dos resultados
print(f"O valor do aumento é: R$ {aumento:.2f}")
print(f"O novo salário é: R$ {novo_salario:.2f}")
```

```
# Entrada de dados
base = float(input("Digite o valor da base: "))
altura = float(input("Digite o valor da altura: "))

# Cálculo da área
area = base * altura

# Resultado
print(f"A área do retângulo é: {area} unidades de medida.")
```

```
# Entrada de dados
distancia = float(input("Digite a distância percorrida (km): "))
combustivel = float(input("Digite a quantidade de combustível gasta (litros): "))

# Cálculo do consumo médio
if combustivel > 0:
    consumo_medio = distancia / combustivel
    print(f"O consumo médio do veículo é de: {consumo_medio:.2f} km/l")
else:
    print("Erro: O combustível gasto deve ser maior que zero.")
```
